实验数据是图像，二进制格式
需要使用pickle读取数据

```python
#二进制读取文件
def load_pickle(f):
    version = platform.python_version_tuple()
    return pickle.load(f,encoding='latin1')
    
def load_batchs(filename):
    with open(filename,'rb') as f:
        dataset = load_pickle(f)
        X = dataset['data']
        Y = dataset['labels']
        #将图像矩阵矩阵分割成1000个32*3*32的矩阵，然后再把对应维的元素兑换1->3 2->1 3->2 
        X = X.reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("float")
        Y = np.array(Y)
    return X,Y

def load_data(path):
    '''
    load all batch data
    '''
    xs = []
    ys = []
    files = os.listdir(path)
    for i in range(1,6):
        filename = os.path.join(path,'data_batch_{}'.format(i))
        X,Y = load_batchs(filename)
        xs.append(X)
        ys.append(Y)
    #np.concatenate  数组、矩阵拼接  默认axis=0,对0轴进行拼接
    X_train = np.concatenate(xs)
    Y_train = np.concatenate(ys)
    #python del 删除的是变量，不删除数据
    del X
    del Y
    X_test,Y_test = load_batchs(os.path.join(path,'test_batch'))
    return X_train,Y_train,X_test,Y_test
```

```python
path = '/Users/yangyang/Desktop/data/cifar-10-batches-py/'
X_train,y_train,X_test,y_test = load_data(path)
```
